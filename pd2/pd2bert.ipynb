### Zadanie 4

Sprawdźmy czy lematyzacja wpływa na jakość tematów.

BERTopic wykorzystuje transformatory, które są oparte na „prawdziwym i czystym” tekście, a nie na tekście bez odrzucanych słów, lematów lub tokenów. Pod koniec obliczeń słowa stop stają się szumem (nie mają charakteru informacyjnego) i mają wartość topic_id = -1. Z tego samego powodu nie należy tokenizować (wewnętrznie) ani lematyzować (nieco subiektywnie) tekstu. To zepsuje tematy.
